```python
# Cell 1: Setup & Load Failures
from failure_reviewer import FailureReviewer
import pandas as pd

reviewer = FailureReviewer()
failed_picks = reviewer.get_failed_picks()

print(f"Found {len(failed_picks)} failed picks")
failed_picks.head()
```

```python
# Cell 2: Review Most Recent Failure
if not failed_picks.empty:
    recent_id = failed_picks.iloc[0]['raw_pick_id']
    print(f"Reviewing most recent failure: ID {recent_id}")

    success = reviewer.fix_and_map(recent_id)
    print("✅ Done!" if success else "❌ Failed")
```

```python
# Cell 3: Bulk Fix Mode (Review Top 5)
top5 = failed_picks.head(5)
print("Top 5 failures:")
for idx, row in top5.iterrows():
    team_raw = str(row.get('team_raw', 'N/A') or 'N/A')
    print(f"  {row['raw_pick_id']}: {team_raw[:40]} [{row['parse_status']}]")

# Interactive: uncomment to fix specific ID
# reviewer.fix_and_map(3)  # Replace 3 with actual ID
```
