```python
# Cell 1: Setup & Run Backtest
from backtest_engine import BacktestEngine
import pandas as pd

engine = BacktestEngine()
results = engine.run_full_backtest()

print('📊 SUMMARY:')
summary = results['summary']
for k, v in summary.items():
    print(f'  {k}: {v}')
```

```python
# Cell 2: Alignment Performance
print('🎯 ALIGNMENT ROI:')
print(results['alignment_stats'].round(3))

print('\n⭐ BEST HANDICAPPERS:')
capper_stats = results['capper_stats']
if not capper_stats.empty:
    print(capper_stats.nlargest(5, 'roi_pct'))
else:
    print('No capper stats available.')
```
